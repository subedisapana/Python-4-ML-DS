```python
import numpy as np

x= np.linspace(0,5,11) #lineary space from 0 to 5 and grap 11 points
y=x ** 2
```

```python
#Basic matplotlib command allows you to quickly create a plot
#Two ways to create matplotlib: Functional and OOP
```

```python
#functional

plt.plot(x, y)

# plt.show()
```

```python
plt.plot(x,y, 'r-') #color
```

```python
plt.plot(x, y)


plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#multiple plots on the same canvas

plt.subplot(1,2,1) # args-: number of rows, colm, plot number
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
```

```python
# Matplot OO method

#Create Figure (empty canvas)
fig = plt.figure() #fig obj created


#Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #passing list with a left, buttom, Width, height

#Plot on that set of axes
axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
#create blank canvas
fig = plt.figure()

#large figure axes 1
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #main axes
axes1.plot(x,y)
axes1.set_title('Larger plot')

#Small figure axes 2
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #insert
axes2.plot(y,x,'r')
axes2.set_title('Smaller plot')
```
