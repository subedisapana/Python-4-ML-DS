```python
import numpy as np

x= np.linspace(0,5,11) #lineary space from 0 to 5 and grap 11 points
y=x ** 2
```

```python
from matplotlib import pyplot as plt
import numpy as np


fig,axes= plt.subplots(nrows=1, ncols=2) #tuple unpacking

# axes.plot(x,y) #works only when we dont pass rows and cols in subplots

# for current_ax in axes:
#     current_ax.plot(x,y)


#axes obj is just a list of  matplot plot axis, the one when you create as fig.add_axes([])
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')

plt.tight_layout() #take care of overlapping plot
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize = (8,2)) #height and inches

# ax = fig.add_axes([0,0,1,1])

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```
