## Subplots()
The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python

fig,axes= plt.subplots(nrows=1, ncols=2) #tuple unpacking

# axes.plot(x,y) #works only when we dont pass rows and cols in subplots

# for current_ax in axes:
#     current_ax.plot(x,y)


#axes obj is just a list of  matplot plot axis, the one when you create as fig.add_axes([])
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')

plt.tight_layout() #take care of overlapping plot
```

## Figure Size and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize = (8,2)) #height and inches

# ax = fig.add_axes([0,0,1,1])

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## Plot Figures

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# ax.plot(x,y, color='orange') #you can also give costume color,  RGB hex code

# ax.plot(x,y, color='orange', lw = 3, alpha=0.5) #alpha determines the transparency of a line, lw is line width

ax.plot(x,y, color='orange', lw = 3, ls = 'dotted') 

#markers: userd when you have few data points

ax.plot(x,y, color='orange', lw = 1, ls = '-', marker='*', markersize = 20,
        markerfacecolor='pink', markeredgewidth=1, markeredgecolor= "blue")



#control over axis appereance

ax.plot(x,y, color='green', lw = 2, ls = 'dotted') 
ax.set_xlim([0,1]) #lower bound and upper bound
ax.set_ylim([0,2])
```
