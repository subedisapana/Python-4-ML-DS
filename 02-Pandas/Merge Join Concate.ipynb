```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':['A0', 'A1'], 'B':['B0','B1'], 'C':['C0', 'C1']}, index = ['ind0','ind1'])
df1
```

```python
df2 = pd.DataFrame({'A':['A2', 'A3'], 'B':['B2','B3'], 'C':['C2', 'C3']}, index = ['ind2','ind3'])
df2
```

```python
df3 = pd.DataFrame({'A':['A4', 'A5'], 'B':['B4','B5'], 'C':['C4', 'C5']}, index = ['ind4','ind5'])
df3
```

```python
##Cancatenation : It basically glues together DataFrames. Note: DImensions should match along the axis you are concatenating
```

```python
pd.concat([df1,df2,df3]) #bydefault axis=0
```

```python
pd.concat([df1,df2,df3], axis=1) #axis=0 , Concat across column
```

```python
import pandas as pd
left = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
                    'A':['A0', 'A1', 'A2', 'A3'], 
                     'B':['B0','B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key':['k0', 'k1', 'k2', 'k3'],
                    'C':['C0', 'C1', 'C2', 'C3'], 
                     'D':['D0','D1', 'D2', 'D3']})
right
```

```python
##Merge : merge dataframes together using similar logic
```

```python
pd.merge(left, right, how='inner', on='key') #Create a row using key column
```

```python
pd.merge(left, right, how='outer', on='key') 
```

```python
pd.merge(left, right, how='right', on='key') 
```

```python
### Joining is a convenient method for combining the columns of two potentially differently-indexed into a single result dataframe
```

```python
left = pd.DataFrame({'A':['A0', 'A1', 'A2', 'A3'], 
                     'B':['B0','B1', 'B2', 'B3']},
                   index =[0,1,2,3])

right = pd.DataFrame({'C':['C0', 'C3', 'C4', 'C5'], 
                     'D':['D0','D3', 'D4', 'D5']},
                    index = [0,3,4,5])
```

```python
left.join(right)
```

```python
left.join(right, how = 'outer')
```
